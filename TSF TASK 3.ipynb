```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize= (10,16))
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
plt.figure(figsize= (15,9))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
corr=df.corr()
corr
```

```python
import seaborn as sns
plt.figure(figsize=(7,6))
sns.heatmap(corr,annot=True,cmap='BrBG')
```

```python
plt.figure(figsize=(20,6))
plt.subplot(2,1,1)
sns.countplot('Region',hue='Category',data=df)
plt.title('DISTRIBUTION OF REGIONS BASED ON CATEGORY')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('REGION',fontsize=12)
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(20,6))
plt.subplot(2,1,2)
sns.countplot('Region',hue='Sub-Category',data=df)
plt.title('DISTRIBUTION OF REGIONS BASED ON SUB CATEGORY')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('REGION',fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
#DISTRIBUTION BASED ON CATEGORY/SUBCATEGORY
plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
sns.countplot('Category',data=df);
plt.title('DISTRIBUTION OF CATEGORY')
plt.xlabel('CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)
```

```python
plt.figure(figsize=(20,10))
sns.countplot('Sub-Category',data=df);
plt.title('DISTRIBUTION OF SUB-CATEGORY')
plt.xlabel('SUB-CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)
plt.show()
```
